```python
import numpy as np
import matplotlib.pyplot as plt

# Define the efficiency calculation function
def efficiency(N, p):
    return N * p * (1 - p) ** (N - 1)

# Set the values for p in steps of 0.05
p_values = np.arange(0, 1, 0.05)

# Calculate efficiencies for N=5 and N=50
efficiency_N5 = [efficiency(5, p) for p in p_values]
efficiency_N50 = [efficiency(50, p) for p in p_values]

# Plot efficiency for N=5
plt.figure(figsize=(10, 8))
plt.plot(p_values, efficiency_N5, label='N=5')
plt.title('Efficiency of Slotted Aloha with N = 5')
plt.xlabel('p')
plt.ylabel('Efficiency')
plt.grid(True)
plt.legend()
plt.show()

# Plot efficiency for N=50
plt.figure(figsize=(10, 8))
plt.plot(p_values, efficiency_N50, label='N=50')
plt.title('Efficiency of Slotted Aloha with N = 50')
plt.xlabel('p')
plt.ylabel('Efficiency')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the efficiency calculation function
def efficiency(N, p):
    return N * p * (1 - p) ** (N - 1)

# Set the range of p values
p_range = np.linspace(0, 0.95, 200)

# Function to plot the efficiency graph for a given N
def plot_efficiency(N):
    plt.figure(figsize=(10, 8))
    for p in np.arange(0, 1, 0.05):
        efficiencies = [efficiency(N, pi) for pi in p_range]
        plt.plot(p_range, efficiencies, label=f'p={p:.2f}')
    plt.title(f'Efficiency of Slotted Aloha with N = {N}')
    plt.xlabel('p')
    plt.ylabel('Efficiency')
    plt.legend(bbox_to_anchor=(1.04,0.5), loc="center left", borderaxespad=0)
    plt.grid(True)
    plt.show()

# Plot efficiency for N=5
plot_efficiency(5)

# Plot efficiency for N=50
plot_efficiency(50)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the efficiency calculation function
def efficiency(N, p):
    return N * p * (1 - p) ** (N - 1)

# Create an array of p values for plotting the curve
p_values = np.linspace(0, 1, 1000)

# Set the ticks for p at which vertical lines will be drawn
p_ticks = np.arange(0, 1, 0.05)

# Plot the efficiency curve and add vertical lines for N=5
plt.figure(figsize=(10, 8))
efficiency_values_N5 = efficiency(5, p_values)
plt.plot(p_values, efficiency_values_N5, label='Efficiency Curve for N=5')
for p in p_ticks:
    plt.axvline(x=p, color='grey', linestyle='--', linewidth=0.5)
plt.title('Efficiency of Slotted Aloha with N = 5')
plt.xlabel('p')
plt.ylabel('Efficiency')
plt.grid(True)
# plt.legend()
plt.show()

# Plot the efficiency curve and add vertical lines for N=50
plt.figure(figsize=(10, 8))
efficiency_values_N50 = efficiency(50, p_values)
plt.plot(p_values, efficiency_values_N50, label='Efficiency Curve for N=50')
for p in p_ticks:
    plt.axvline(x=p, color='grey', linestyle='--', linewidth=0.5)
plt.title('Efficiency of Slotted Aloha with N = 50')
plt.xlabel('p')
plt.ylabel('Efficiency')
plt.grid(True)
# plt.legend()
plt.show()
```
